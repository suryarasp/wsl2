```python
import yfinance as yf
import pandas as pd

# Define the ticker symbols for NSE200 stocks
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Read the CSV file containing the symbols
import pandas as pd
# Define the path to the CSV file
csv_file_path = '/content/drive/My Drive/nse200_symbols.csv'

# Read the CSV file
df_symbols =pd.read_csv('/content/drive/MyDrive/Colab Notebooks/input/ind_nifty200list.csv',index_col=False)

# Extract the symbols as a list


# Extract the symbols as a list
nse200_tickers=df_symbols.iloc[:,0].tolist()




nse200_tickers1 = [
    'RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS', 'ICICIBANK.NS',
    'HINDUNILVR.NS', 'HDFC.NS', 'SBIN.NS', 'KOTAKBANK.NS', 'BHARTIARTL.NS',
    'ITC.NS', 'ASIANPAINT.NS', 'BAJFINANCE.NS', 'LT.NS', 'HCLTECH.NS',
    'AXISBANK.NS', 'MARUTI.NS', 'SUNPHARMA.NS', 'NESTLEIND.NS', 'ULTRACEMCO.NS',
    'TITAN.NS', 'M&M.NS', 'INDUSINDBK.NS', 'WIPRO.NS', 'POWERGRID.NS',
    'NTPC.NS', 'HDFCLIFE.NS', 'BAJAJFINSV.NS', 'TECHM.NS', 'JSWSTEEL.NS',
    'DIVISLAB.NS', 'TATASTEEL.NS', 'HEROMOTOCO.NS', 'BPCL.NS', 'ADANIENT.NS',
    'SHREECEM.NS', 'GRASIM.NS', 'BRITANNIA.NS', 'CIPLA.NS', 'ONGC.NS',
    'TATAMOTORS.NS', 'COALINDIA.NS', 'ADANIPORTS.NS', 'SBILIFE.NS', 'APOLLOHOSP.NS',
    'BAJAJ-AUTO.NS', 'EICHERMOT.NS', 'UPL.NS', 'HINDALCO.NS', 'DRREDDY.NS'
    # Add more tickers to complete the NSE200 list
]
# Create a DataFrame to store the stock data
stock_data = []

# Fetch the data for each ticker
for ticker in nse200_tickers:
    stock = yf.Ticker(ticker)
    data = stock.history(period="5d")
    if len(data) < 2:
        continue
    closing_price = data['Close'].iloc[-1]
    prev_closing_price = data['Close'].iloc[-2]
    volume = data['Volume'].iloc[-1]
    prev_volume = data['Volume'].iloc[-2]
    volume_change = ((volume - prev_volume) / prev_volume) * 100
    price_change = ((closing_price - prev_closing_price) / prev_closing_price) * 100
    stock_data.append({
        'Symbol': ticker,
        'Closing Price': closing_price,
        'Prev Closing Price': prev_closing_price,
        'Volume': volume,
        'Prev Volume': prev_volume,
        'Volume Change (%)': volume_change,
        'Price Change (%)': price_change
    })

# Convert to DataFrame
df = pd.DataFrame(stock_data)

# Function to find top 5 price increased stocks
def get_top_5_price_increased(df):
    top_5 = df.nlargest(5, 'Price Change (%)')
    return top_5

# Get top 5 price increased stocks
top_5_stocks = get_top_5_price_increased(df)

# Display top 5 stocks
print(top_5_stocks)

# Fetch the past 10 days price and volume data for the top 5 stocks
top_5_tickers = top_5_stocks['Symbol'].tolist()
top_5_data = {}

for ticker in top_5_tickers:
    stock = yf.Ticker(ticker)
    data = stock.history(period="5d")
    top_5_data[ticker] = data

# Plotting the data using Plotly
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Create subplots
fig = make_subplots(rows=5, cols=2, subplot_titles=top_5_tickers,
                    shared_xaxes=True, vertical_spacing=0.1)

row = 1
for ticker in top_5_tickers:
    data = top_5_data[ticker]
    # Plot closing prices
    fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name=f'{ticker} Close'), row=row, col=1)
    # Plot volumes
    fig.add_trace(go.Bar(x=data.index, y=data['Volume'], name=f'{ticker} Volume'), row=row, col=2)
    row += 1

# Update layout
fig.update_layout(height=1500, width=1000, title_text="Top 5 Stocks Price and Volume Over the Last 5 Days")
fig.update_yaxes(title_text="Price", row=1, col=1)
fig.update_yaxes(title_text="Volume", row=1, col=2)

# Show plot
fig.show()
```
